The data is recorded at 5-minute intervals.
Key features include:

CGM (Continuous Glucose Monitoring) values
Finger stick blood glucose measurements (sparse)
Basal insulin rate
Heart rate (hr)
Galvanic skin response (gsr)
Carbohydrate input
Bolus insulin injections


The 'cbg' column represents the continuous glucose monitoring values, which is our target for prediction.
